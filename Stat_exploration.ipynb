```python
import pandas as pd
import os
import glob
import numpy as np
import ast
import matplotlib.pyplot as plt
```

```python
path = os.getcwd()
lfile = glob.glob(f'{path}/Stat_store/*50.csv')
df = pd.read_csv(lfile[0])
df.History = [ast.literal_eval(i) for i in df.History]
```

```python
df1 = df.loc[df.Id_joueur == 1]
df2 = df.loc[df.Id_joueur == 2]
Max_score1 = df1.Max_Score
Max_score2 = df2.Max_Score
Max_score_l = [Max_score1, Max_score2]
```

```python
for i in range(len(Max_score_l)):
    plt.hist(Max_score_l[i], label = f'Player nÂ°{i+1}')
plt.legend()
plt.xlabel('Score max')
plt.ylabel()
```
