```python
import pandas as pd
import os
import glob
import numpy as np
import ast
import matplotlib.pyplot as plt
path = os.getcwd()
```

```python
def plot_hist_max_value(path, Title):
    lfile = glob.glob(path)
    df = pd.read_csv(lfile[0])
    df.History = [ast.literal_eval(i) for i in df.History]
    df1 = df.loc[df.Id_joueur == 1]
    df2 = df.loc[df.Id_joueur == 2]
    Max_score1 = df1.Max_Score
    Max_score2 = df2.Max_Score
    Max_score_l = [Max_score1, Max_score2]
    for i in range(len(Max_score_l)):
        plt.hist(Max_score_l[i], label = f'Player nÂ°{i+1}', alpha=0.5)
    plt.legend()
    plt.xlabel('Score max')
    plt.ylabel('Occurence')
    plt.title(Title)
```

```python
def plot_hist_winner(path, Title):
    
    lfile = glob.glob(path)
    df = pd.read_csv(lfile[0])
    df.History = [ast.literal_eval(i) for i in df.History]
    df1 = df.loc[df.Id_joueur == 1]
    df2 = df.loc[df.Id_joueur == 2]
    Max_score1 = df1.Max_Score
    Max_score2 = df2.Max_Score
    Winner = []
    for j1, j2 in zip(Max_score1, Max_score2):
        if j1>j2:
            Winner.append(1)
        elif j2>j1:
            Winner.append(2)
        else:
            Winner.append(0)
    plt.hist(Winner, histtype='bar')
    plt.title(Title)
```

```python
size = 50
plot_hist_max_value(f'{path}/Stat_store/*{size}.csv', Title=f'Comparison for sample size = {size}')
```

```python
size = 50
plot_hist_winner(f'{path}/Stat_store/*{size}.csv', Title=f'Comparison for sample size = {size}')
```
